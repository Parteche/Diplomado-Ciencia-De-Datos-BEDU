```python
#Datos faltantes
#Verificar valores nulos en todas las columnas:
print(df.isnull().sum())
#El resultado nos arroja que la columna 'Book Name' tiene 23 registros nulos y la columna language_code tiene 53
```

```python
#Eliminando registros nulos
df.dropna(inplace=True)
#Verificamos de nuevo registros nulos
print(df.isnull().sum())
#Los resultados son favorables, cada vez tenemos una base de datos mas limpia
```

```python
#Registros repetidos:
print(df.duplicated().sum())
#El resultado nos indica que no tenemos datos duplicados en nuestra base de datos
```

```python
#Columnas irrelevantes
#Columnas categoricas
columns_to_check_cat = ['Book Name', 'Author', 'language_code', 'Author_Rating', 'genre', 'Publisher ']
```

```python
#Si alguna columna categorica nos da un valor igual a 1 esta se considera irrelevante, significa que todos los valores
#son iguales
unique_counts = df[columns_to_check_cat].nunique()
print(unique_counts)
```

```python
#Otro criterio para eliminar datos irrelevantes es si no nos sirven para contestar nuestras preguntas, por ejemplo la columna
#language_code, todos los libros estan en ingles, no hace la diferencia esa la informacion de esa columna
#Primero exploraremos la columna language_code:
df['language_code'].value_counts()
#Podemos observar que 4 de los diferentes lenguages son el mismo solo que con error tipografico
#y los otros 4 que en total representan 9 filas son insignificantes para nuestras base de datos, por lo tanto esta columna no nos sirve realmente
#para las consultas que queremos hacer
```

```python
#Segundo, explorando la columna Author_Rating:
df['Author_Rating'].value_counts()
#A pesar de que son pocas categorias, si nos podrian servir para nuestras consultas y ademas ninguna se repite
```

```python
#Tercero, explorando la columna genre:
df['genre'].value_counts()
#De los 4 diferentes generos que nos arroja, dos significan los mismo, en este caso vamos a unificar esos dos generos para tener solo 3 en total
#No eliminaremos esta columna por que creo que podemos necesitar estos datos
```

```python
#Primero hacemos un diccionario con los diferentes subniveles de genero y con el nombre final que queremos que tenga:
unificando_genero = {
    'genre fiction': 'fiction', 'fiction': 'fiction', 'nonfiction': 'no fiction', 'children': 'children'
}

#Codigo para remplazar nombres (aqui usamos el diccionario que hicimos para que reemplace los registros mal nombrados):
#Usamos en metodo de pandas .raplace(dicccionario)
df['genre'] = df['genre'].replace(unificando_genero)
print("DataFrame after replacement:")
df.head()
```

```python
#Eliminando columnas irrelevantes: language_code
#Usaremos el metodo de pandas .drop()
df = df.drop('language_code', axis=1)
df.head()
```

```python
#Nos hemos dado cuenta que la columna publisher nos da error, y tiene datos irrelevantes asi que la vamos a eliminar, pero primero la examinaremos
#df['Publisher'].value_counts() no podemos verla, nos marca error
#df = df.drop('Publisher', axis=1)
#df.head()
#La columna Publisher esta maldita...
df['Publisher '].value_counts()
#Ya encontramos el error, en el nombre tiene un espacio al final que obvio no escribiamos cuando usabamos el nombre de la columna
```

```python
#Cambiando el nombre de la columna Publisher para corregir el espacio en blanco:
#Usamos el metodo .rename()
df.rename(columns={'Publisher ': 'Editorial'}, inplace=True)
df.head()
```

```python
#Unificaremos algunos nombres repetidos de editoriales:
unificando_editorial = {
    'Random House LLC': 'Penguin Random House', 'Penguin Group (USA) LLC': 'Penguin Random House', 'Amazon Digital Services,  Inc.': 'Amazon KDP', 'Simon and Schuster Digital Sales Inc': 'Simon and Schuter',
    'HarperCollins Christian Publishing ': 'HarperCollins Publishing'
}

#Usamos en metodo de pandas .raplace(dicccionario)
df['Editorial'] = df['Editorial'].replace(unificando_editorial)
print("DataFrame after replacement:")
df.head()
```

```python
#Columnas numericas
#Para saber si una columna numerica es irrelevante checamos su desviacion estandar, si esta es igual a cero
#quiere decir que los valores en esa columna son iguales, no hay variacion de numeros, osea es irrelevante
df.describe()
#El metodo .describe() por default toma solo las columnas de valor numerico
#No hay columnas numericas con desviacion estandar igual a cero
#Los valores repetidos en las columnas numericas si pueden ser
```
